```python
import sklearn
import pandas as pd
from matplotlib import pyplot as plt
    
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
d = pd.read_csv("../data/doped_stats.csv")
```

```python
d.head()
```

```python
y = d["Target"]
X = d.drop(columns=["Target", "Element"])
```

```python
tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X, y)
```

```python
f, ax = plt.subplots(1, 1, figsize=(12, 7))
plot_tree(tree, feature_names=X.columns, class_names=["Replaces Fe", "Replaces Rh"], ax=ax)
plt.show()
```
